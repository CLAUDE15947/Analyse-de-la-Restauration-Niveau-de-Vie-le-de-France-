```python
import pandas as pd
import unicodedata
import re

# Charger le fichier
df = pd.read_csv("restaurants_paris_enriched_sample.csv")

# fonction de nettoyage
def clean_text(value):
    if pd.isna(value):
        return value

    # Normalisation Unicode
    value = unicodedata.normalize('NFKD', str(value)).encode('ascii', 'ignore').decode('utf-8')
    # Suppression caractères spéciaux
    value = re.sub(r"[^A-Za-z0-9 ,.\-]", "", value)
    # Suppression espaces multiples et strip
    value = re.sub(r"\s+", " ", value)

    return value.strip()

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(clean_text)

# Supprimer les restaurats hors de paris
df = df[df['zone_paris'].str.strip() != 'horsparis']

# 4. Suppressions de colonnes
df = df.dropna(axis=1, how='all')
df = df.drop(columns=["lat_round", "lon_round", "latitude", "longitude"], errors='ignore')

# Supprimer les lignes avec valeurs manquantes
df = df.dropna()

# Réinitialiser l'index
df = df.reset_index(drop=True)


# Sauvegarder
df.to_csv("restaurants_paris_clean.csv", index=False)
```
