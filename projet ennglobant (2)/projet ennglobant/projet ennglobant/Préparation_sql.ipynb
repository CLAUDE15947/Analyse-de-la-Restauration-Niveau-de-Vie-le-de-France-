```python
import pandas as pd
import unicodedata
import re
import csv

df = pd.read_csv("restaurants_agg.csv")

# RENOMMAGE DES COLONNES
mapping_colonnes = {
    'nom': 'nom_restaurant',
    'prix_intervalle': 'prix_moyen',
    'NOM': 'nom_arrondissement',
    'INSEE_FROM_CP': 'insee_from_cp',
    'INSEE_COM': 'insee_com',
    'POP': 'pop_totale',
    'Population_3_20': 'pop_3_20_ans',
    'C20_FAM': 'nb_familles',
    'C20_FAMMONO': 'nb_familles_mono',
    'QL3_20': 'part_pop_3_20',
    'QLFMONO': 'part_fam_mono',
    'RDISP_MED': 'revenu_median',
    'tx_BAC': 'taux_bac',
    'tx_O': 'taux_ouvrier',
    'tx_Chom': 'taux_chomage',
    'FDep': 'facteur_deprivation',
    'APL': 'part_apl'
}
df = df.rename(columns=mapping_colonnes)

# Ajout de l'ID unique en première position
df.insert(0, 'id_restaurant', range(1, 1 + len(df)))

# Code postal : toujours 5 caractères
df['code_postal'] = df['code_postal'].astype(str).apply(lambda x: x.split('.')[0]).str.zfill(5)

# Arrondissement : entier
df['arrondissement'] = df['arrondissement'].fillna(0).astype(int)

# INSEE codes : chaines de caractères propres
for col in ['insee_from_cp', 'insee_com']:
    df[col] = df[col].astype(str).apply(lambda x: x.split('.')[0])

# Sélection finale des colonnes dans le bon ordre
cols_finales = [
    'id_restaurant', 'nom_restaurant', 'prix_moyen', 'rating', 'nb_notes',
    'adresse', 'ville', 'main_category', 'commune', 'code_postal',
    'arrondissement', 'zone_paris', 'insee_from_cp', 'insee_com',
    'nom_arrondissement', 'pop_totale', 'pop_3_20_ans', 'nb_familles',
    'nb_familles_mono', 'part_pop_3_20', 'part_fam_mono', 'revenu_median',
    'taux_bac', 'taux_ouvrier', 'taux_chomage', 'facteur_deprivation', 'part_apl'
]
# On ne garde que les colonnes qui existent (au cas où)
cols_existantes = [c for c in cols_finales if c in df.columns]
df = df[cols_existantes]

# EXPORT
fichier_sortie = "restaurants_paris_sql_ready.csv"
df.to_csv(
    fichier_sortie,
    index=False,
    quoting=csv.QUOTE_NONNUMERIC,
    quotechar='"',
    encoding='utf-8'
)
```
