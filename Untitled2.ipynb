```python
import zipfile
import os
from torchvision import datasets, transforms
from torch.utils.data import random_split

# Décompression du fichier local
zip_path = "/content/archive (12).zip"
extract_path = "/content/pokemon_data"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Configuration PyTorch (Page 11 du projet)
transform = transforms.Compose([
    transforms.Resize((64, 64)), # [cite: 38, 87]
    transforms.Grayscale(num_output_channels=1), # [cite: 38, 87]
    transforms.ToTensor(), # [cite: 39, 87]
])

# Chargement. Note : datasets.ImageFolder cherche des sous-dossiers par classe
# Si l'extraction crée un dossier "PokemonData", pointez vers lui.
dataset = datasets.ImageFolder(root=extract_path, transform=transform) # [cite: 87]

# Division 70% Train / 15% Val / 15% Test
train_size = int(0.7 * len(dataset)) # [cite: 87]
val_size = int(0.15 * len(dataset)) # [cite: 87]
test_size = len(dataset) - train_size - val_size # [cite: 87]

train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size]) # [cite: 87]
print(f"Dataset prêt : {len(dataset)} images trouvées.")
```

```python
import numpy as np

def to_numpy(dataset_split):
    X, y = [], []
    for img, label in dataset_split:
        # On aplatit l'image (Flatten) : 64x64 devient un vecteur de 4096 [cite: 41]
        X.append(img.numpy().flatten())
        y.append(label)
    return np.array(X), np.array(y)

print("Conversion des données en cours...")
X_train, y_train = to_numpy(train_ds)
X_val, y_val = to_numpy(val_ds)
X_test, y_test = to_numpy(test_ds)
print(f"Format final X_train : {X_train.shape}")
```

```python
from sklearn.decomposition import PCA

# Réduction à 150 composantes pour optimiser les performances [cite: 89]
pca = PCA(n_components=150)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

print("PCA terminée. Données prêtes pour le SVM.")
```

```python
import os
from torchvision import datasets

# Exploration du dossier extrait pour trouver le dossier contenant les classes
base_path = "/content/pokemon_data"
subdirs = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

# Si le dossier extrait contient un seul sous-dossier, on descend d'un niveau
if len(subdirs) == 1:
    correct_path = os.path.join(base_path, subdirs[0])
else:
    correct_path = base_path

print(f"Nouveau chemin utilisé : {correct_path}")

# Rechargez le dataset avec le chemin corrigé
dataset = datasets.ImageFolder(root=correct_path, transform=transform)
print(f"Classes trouvées avec le nouveau chemin : {len(dataset.classes)}")
```

```python
import numpy as np
from sklearn.decomposition import PCA
from torch.utils.data import random_split

# 1. Séparation Train/Val/Test (70% / 15% / 15%) [cite: 46, 87]
train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

# 2. Fonction de conversion vers Numpy (Flatten) [cite: 41, 88]
def to_numpy(ds):
    X, y = [], []
    for img, label in ds:
        X.append(img.numpy().flatten()) # Aplatit l'image 64x64 en vecteur de 4096
        y.append(label)
    return np.array(X), np.array(y)

X_train, y_train = to_numpy(train_ds)
X_val, y_val = to_numpy(val_ds)
X_test, y_test = to_numpy(test_ds)

# 3. Extraction de features via PCA (150 composantes recommandées) [cite: 42, 89]
pca = PCA(n_components=150)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

print("Extraction de features terminée. Données prêtes pour le SVM.")
```

```python
from sklearn.svm import SVC

# Initialisation du SVM avec noyau RBF [cite: 32, 47, 90]
svm = SVC(kernel='rbf', C=20, gamma='scale')

print(f"Entraînement en cours sur {len(np.unique(y_train))} classes...")
# Le modèle apprend à partir des caractéristiques extraites par la PCA [cite: 89, 90]
svm.fit(X_train_pca, y_train)

# Calcul de la précision sur le set de validation pour vérifier la performance [cite: 49, 90]
val_acc = svm.score(X_val_pca, y_val)
print(f"Succès ! Précision sur validation : {val_acc:.2%}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Prédiction sur l'ensemble de test (données inconnues du modèle) [cite: 46, 91]
y_pred = svm.predict(X_test_pca)

# Génération du rapport de performance détaillé [cite: 51, 91]
print("\n--- Rapport de Classification ---")
print(classification_report(y_test, y_pred))

# Affichage de la matrice de confusion sous forme de Heatmap [cite: 52, 91]
plt.figure(figsize=(12, 10))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=False, cmap="Blues")
plt.title("Matrice de Confusion : 150 Classes Pokémon")
plt.xlabel("Classes Prédites")
plt.ylabel("Classes Réelles")
plt.show()
```
